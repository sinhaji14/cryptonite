```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('new.csv')
```

```python
x=data.x.to_numpy()
y=data.y.to_numpy()
```

```python
plt.scatter(x,y)
```

```python
x = x/500
y= y/500
```

```python
m,c= np.random.random(),np.random.random()
```

```python
def learn(x, y, m, c, epoch):
    learning_rate = .01
    for i in range(epoch):
        error = y - (m*x + c)
        x_error = (y - (m*x + c))*x
        error = np.sum(error)/len(x)
        x_error = np.sum(x_error)/len(x)
        delta_m = learning_rate*x_error
        delta_c = learning_rate*error
        m += delta_m
        c += delta_c
    return m, c

```

```python
m, c = 0.0, 0.0
m, c = learn(x, y, m, c, 20000)
```

```python
x1 = [0,1]
y1 = [0*m+c,1*m+c]
```

```python
plt.scatter(x,y)
plt.plot(x1,y1)
```
