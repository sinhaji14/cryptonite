```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Heart.csv")
```

```python
df
```

```python
df = df.drop(columns = 'Unnamed: 0') #used to remove the column
df
```

```python
df['ChestPain'] = df['ChestPain'].astype('category') # used to change letter type to number
df['ChestPain'] =df['ChestPain'].cat.codes
df['Thal'] = df['Thal'].astype('category') # used to change letter type to number
df['Thal'] =df['Thal'].cat.codes
df['AHD'] = df['AHD'].astype('category') # used to change letter type to number
df['AHD'] =df['AHD'].cat.codes
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna() #dropping null values to make data simpler
df
```

```python
X= df.drop(columns='AHD') #depedent variable
X
```

```python
y= df['AHD']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=.3 ,random_state=21) #rs is uded generate the same train set split if necessary
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

```

```python
X_train_scaled = scaler.fit_transform(X_train) #fit helps in calculating mean and standard deviation
X_test_scaled = scaler.transform(X_test)   #transform helps in calculating z, z=(x-u)/s
# we dont want our model to know about our test data , test data is for evaluation purpose onli.
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(random_state=0).fit(X_train_scaled,y_train)
```

```python
log_reg.predict(X_train_scaled)
```

```python
log_reg.score(X_train_scaled,y_train)
```

```python
log_reg.score(X_test_scaled,y_test)
```

```python
log_reg1 = LogisticRegression(random_state=0,
                              C=1, fit_intercept=True
                              ).fit(X_train_scaled,y_train)
```

```python
log_reg1.score(X_train_scaled,y_train)
```
